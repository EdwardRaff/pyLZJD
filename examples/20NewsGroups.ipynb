```python
from pyLZJD import digest, sim, vectorize
import numpy as np
from sklearn.datasets import fetch_20newsgroups
from sklearn import metrics
from sklearn.linear_model import LogisticRegression, SGDClassifier

from multiprocessing import Pool
import tqdm
```

In this example, our goal is to show that LZJD can be used on string/textual data as well. It works best when the text is longer, which is true fo the 20-news groups dataset. While a TfidfVectorizer will work better, LZJD is still decent when only a few classes are present. 

```python
categories = ['alt.atheism', 'comp.graphics']
```

```python
newsgroups_train = fetch_20newsgroups(subset='train', categories=categories, remove=('headers', 'footers', 'quotes'))
newsgroups_test = fetch_20newsgroups(subset='test', categories=categories, remove=('headers', 'footers', 'quotes'))
```

```python
#What does our data look like?
print(newsgroups_train.data[7])
```

```python
X_train = vectorize(newsgroups_train.data)
X_test = vectorize(newsgroups_test.data)
```

```python
clf = LogisticRegression(multi_class='multinomial', solver='saga')
clf.fit(X_train, newsgroups_train.target)
```

```python
pred = clf.predict(X_test)
metrics.f1_score(newsgroups_test.target, pred, average='macro')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer()
X_train_tfidf = vectorizer.fit_transform(newsgroups_train.data)
X_test_tfidf = vectorizer.transform(newsgroups_test.data)
```

```python
clf = LogisticRegression()
clf.fit(X_train_tfidf, newsgroups_train.target)
pred = clf.predict(X_test_tfidf)
metrics.f1_score(newsgroups_test.target, pred, average='macro')
```
